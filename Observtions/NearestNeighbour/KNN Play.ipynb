```python
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
import networkx as nx
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def make() :
    n = random.randint(25,1000)
    m = random.randint(2,7)
    g = nx.barabasi_albert_graph(n,m)
    kmax = max(g.degree())
    edge = len(g.edges())
    degree_count = dict()
    # DICTIONARY THAT KEEPS COUNT OF HOW MANY VERTICES ARE THERE OF EACH DEGREE
    for i in xrange(n) :
        degree_count[g.degree(i)] = degree_count.get(g.degree(i), 0) + 1
    rank = dict()
    su = 1
    for k in sorted(degree_count.keys(),reverse = True) :
        rank[k] = su
        su += degree_count[k]
    deg = g.degree()
    train(n,edge,kmax,rank,deg)
```

```python
def train(n,edge,kmax,rank,deg) :
    f = open("data.txt","a")
    for i in xrange(n) :
        s = '{0},{1},{2},{3}\n'.format(str(n),str(edge),str(kmax),str(rank[deg[i]]))
        f.write(s)
    f.close()
```

```python
for i in xrange(5) :
    make()
```

```python
neigh = KNeighborsRegressor(n_neighbors=25,weights="distance")
```
